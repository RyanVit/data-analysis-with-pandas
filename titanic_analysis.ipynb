# Análise Titanic

#### importando bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Configuração do estilo dos gráficos

```python
sns.set(style="whitegrid")
```

#### Carregando os dados do arquivo CSV

```python
df = pd.read_csv('titanic.csv')
```

#### Verificar as primeiras linhas do DataFrame

```python
print(df.head())
```

#### Verificar as colunas do DataFrame

```python
print(df.columns)
```

#### Informações gerais sobre os dados

```python
print(df.info())
```

#### Estatísticas descritivas dos dados

```python
print(df.describe())
```

#### Verificar se a coluna 'Survived' está presente

```python
if 'Survived' in df.columns:
    print("A coluna 'Survived' está presente.")
else:
    print("A coluna 'Survived' não está presente.")
```

#### Contagem de sobreviventes


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Contagem de Sobreviventes')
plt.xlabel('Sobreviveu (0 = Não, 1 = Sim)')
plt.ylabel('Contagem')
plt.show()
```

#### Sobrevivência por classe

```python
plt.figure(figsize=(30, 10))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Sobrevivência por Classe')
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()
```

#### Distribuição de idade dos sobreviventes

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 1]['Age'].dropna(), kde=True, color='green', bins=30)
plt.title('Distribuição de Idade dos Sobreviventes')
plt.xlabel('Idade')
plt.ylabel('Densidade')
plt.show()
```

#### Relação entre idade e tarifa

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette={0: 'red', 1: 'blue'}, hue_order=[0, 1])
plt.title('Relação entre Idade e Tarifa')
plt.xlabel('Idade')
plt.ylabel('Tarifa')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()
```
